```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python

x=DFtrain.drop("price_range",axis=1)
y=DFtrain['price_range']
```

```python
x[:1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.columns
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_squared_error
import math
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
ypred=[]
for i in range(len(y_pred)):
    ypred.append(int(y_pred[i]))
acc=accuracy_score(y_test,ypred)
mse=mean_squared_error(y_test,ypred)
rmse=math.sqrt(mse)
print(acc,mse,rmse)
np=reg.predict(DFtest.drop(['id'],axis=1))
print(np)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,mean_squared_error
import math
reg=LogisticRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
ypred=[]
for i in range(len(y_pred)):
    ypred.append(int(y_pred[i]))
acc=accuracy_score(y_test,ypred)
mse=mean_squared_error(y_test,ypred)
rmse=math.sqrt(mse)
print(acc,mse,rmse)
np=reg.predict(DFtest.drop(['id'],axis=1))
for i in range(len(np)):
    print(np[i]) 
```

```python
print("using decision tree classifier using gini impurity(cart)")
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,mean_squared_error
import math
reg=DecisionTreeClassifier(criterion="entropy")
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
ypred=[]
for i in range(len(y_pred)):
    ypred.append(int(y_pred[i]))
acc=accuracy_score(y_test,ypred)
mse=mean_squared_error(y_test,ypred)
rmse=math.sqrt(mse)
print(acc,mse,rmse)
np=reg.predict(DFtest.drop(['id'],axis=1))
np 
```
